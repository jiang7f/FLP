```python
import pandas as pd
import ast

# 
data = pd.read_csv('evaluate.csv')

#  best_solution_probs 
def process_value(value):
    try:
        # 
        value = ast.literal_eval(value)
    except (ValueError, SyntaxError):
        # 
        pass
    
    # 
    if isinstance(value, list):
        return max(value)
    else:
        return value

#  .loc  DataFrame 
data.loc[:, 'best_solution_probs'] = data['best_solution_probs'].apply(process_value)

#  CSV 
# data.to_csv('evaluate_processed.csv', index=False)
```

```python
data = data[["pkid", "strategy", "best_solution_probs"]]
data
```

```python
data = data[data['best_solution_probs'] != 'timeout']
data = data[data['strategy'] < 4]
```

```python
data = data.groupby(['pkid', 'strategy']).mean().reset_index()
data
```

```python
dataused = data.pivot(index ='strategy', columns='pkid', values=["best_solution_probs"])
dataused
```
