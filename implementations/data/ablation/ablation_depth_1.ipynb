```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
scale = 1.05
fig= plt.figure(figsize=(25*scale, 12*scale))
mpl.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
    'font.family': 'Times New Roman',
    'font.size': 65,
    'axes.unicode_minus': False,
    'mathtext.fontset': 'custom',
    'mathtext.rm': 'Times New Roman',
    'mathtext.it': 'Times New Roman:italic',
    'mathtext.bf': 'Times New Roman:bold',
    'axes.linewidth': 5,
    'xtick.major.size': 20,
    'xtick.major.width': 5,
    'ytick.minor.size': 10,
    'ytick.minor.width': 3,
    'ytick.major.size': 20,
    'ytick.major.width': 5,
    "markers.fillstyle" :  'left' , ## default
    "lines.markersize" : 30,
    "lines.markeredgewidth" : 3,
    # 'lines.markeredgecolor': 'black',
})
bar_width = 0.118
shrink = 0.05

colors = ["#7CBEEC", "#7EB57E", "#FF6F45", "#3274A1"]
labels = ['depth', 'rss usage (MB)']
# for i, metric in enumerate( ["culled_depth",'rss_usage']):
i = 0
axes = plt.axes([0, 0, 1, 0.37])
ylabel = 'depth'

indexs = [i for i in np.arange(0, len(dataused.index.values))]
for i in range(0, len(indexs) - 1, 3):
    indexs[i] += shrink
    indexs[i + 2] -= shrink

ymetric = "culled_depth" # best_solution_probs
axes.bar([index - 5 * bar_width / 2 for index in indexs], dataused[ymetric]['[0, 0]'].apply(lambda x: float(x) if x != 'timeout' else np.nan), linewidth=3, edgecolor='black', width=bar_width, label="strategy 1",
        color=colors[0])
axes.bar([index - 3 * bar_width / 2 for index in indexs], dataused[ymetric]['[0, 1]'].apply(lambda x: float(x) if x != 'timeout' else np.nan), linewidth=3, edgecolor='black', width=bar_width, label="strategy 2",
        color=colors[1])
axes.bar([index - 1 * bar_width / 2 for index in indexs], dataused[ymetric]['[0, 2]'].apply(lambda x: float(x) if x != 'timeout' else np.nan), linewidth=3, edgecolor='black', width=bar_width, label="strategy 3",
        color=colors[2])
axes.bar([index + 1 * bar_width / 2 for index in indexs], dataused[ymetric]['[1, 0]'].apply(lambda x: float(x) if x != 'timeout' else np.nan), linewidth=3, edgecolor='black', width=bar_width, label="strategy 4",
        color=colors[3])
axes.bar([index + 3 * bar_width / 2 for index in indexs], dataused[ymetric]['[1, 1]'].apply(lambda x: float(x) if x != 'timeout' else np.nan), linewidth=3, edgecolor='black', width=bar_width, label="strategy 5",
        color=colors[2])
axes.bar([index + 5 * bar_width / 2 for index in indexs], dataused[ymetric]['[1, 2]'].apply(lambda x: float(x) if x != 'timeout' else np.nan), linewidth=3, edgecolor='black', width=bar_width, label="strategy 6",
        color=colors[1])
index = indexs
axes.set_xticks(index, minor=False)
# axes.set_yticks(range(0, 3500, 1000), minor=False)
axes.set_ylabel(r"depth")
axes.grid(axis="y", color="gray", linestyle="--", linewidth=1, dashes=(2, 2))
axes.set_yscale("log")
axes.set_xticklabels(range(0, 9), minor=False)
axes.set_xlim(0 - 4 * bar_width + shrink, 8 + 4 * bar_width - shrink)
axes.legend(frameon=False, bbox_to_anchor=(0, 1.05), loc='lower left', ncol=3, prop={'size': 60})
```
