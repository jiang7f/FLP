```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
data = pd.read_csv('real_qpu_data.csv')
data.drop(columns=['pbid'], inplace=True)
data

```

```python
latency_df = pd.read_csv('_latency&transpile_time.csv')

latency_df
```

```python
qpu_data =data.merge(latency_df, on=['pkid', 'layers', 'method', 'backend'])
qpu_data['backend'] = qpu_data['backend'].str.replace('^ibm_', '', regex=True)
qpu_data
```

```python
qpu_data['runtime'] = qpu_data.latency * qpu_data.shots * qpu_data.iteration_count / 10 ** 9 + qpu_data.transpile_time
qpu_data
```

```python
probs = ['F1','G1','K1']
qpu_data.pkid = qpu_data.pkid.apply(lambda x: probs[x])
qpu_data
```

```python
# 设置索引列
index_columns = ['pkid', 'layers', 'method', 'backend']
value_columns = [col for col in qpu_data.columns if col not in index_columns]

# 计算每组的均值
grouped_means = qpu_data.groupby(['layers', 'method'])[value_columns].mean().reset_index()
grouped_means['pkid'] = 'Avg'
grouped_means['backend'] = 'Avg'
# 添加均值行到 DataFrame
qpu_data = pd.concat([qpu_data, grouped_means], ignore_index=True)

qpu_data
```

```python
data2draw = qpu_data.pivot_table(index =['backend','pkid','shots'], columns='method', values=["ARG",'in_constraints_probs','best_solution_probs','iteration_count','latency','transpile_time','runtime'], sort=False)
data2draw
```

```python
data2draw.groupby(['backend','shots']).mean()
```

```python
data1024shots = data2draw.xs(1024, level='shots')
```
