```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
data = pd.read_csv('ibm_data.csv')
data.drop(columns=['pbid'], inplace=True)
data
```

```python
latency_df = pd.read_csv('_latency&transpile_time.csv')
latency_df
```

```python
qpu_data =data.merge(latency_df, on=['pkid', 'layers', 'method', 'backend'])
qpu_data
```

```python
qpu_data['runtime'] = qpu_data.latency * qpu_data.shots * qpu_data.iteration_count / 10 ** 9 + qpu_data.transpile_time
qpu_data
```

```python
probs = ['F1','G1','K1']
qpu_data.pkid = qpu_data.pkid.apply(lambda x: probs[x])
qpu_data
```

```python
# 
index_columns = ['pkid', 'layers', 'method', 'backend']
value_columns = [col for col in qpu_data.columns if col not in index_columns]

# 
grouped_means = qpu_data.groupby(['layers', 'method'])[value_columns].mean().reset_index()
grouped_means['pkid'] = 'Avg'
#  DataFrame
df_combined = pd.concat([qpu_data, grouped_means], ignore_index=True)

qpu_data = df_combined
for row in qpu_data[index_columns].values:
    print(row)
```

```python
data2draw = qpu_data.pivot(index =['backend','pkid','shots'], columns='method', values=["ARG",'in_constraints_probs','best_solution_probs','iteration_count','latency','transpile_time','runtime'])
data2draw
```

```python
data2draw.groupby(['backend','shots']).mean()
```

```python
data1024shots = data2draw.xs(1024, level='shots')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Initialize the figure and axis settings
scale = 1.05
fig = plt.figure(figsize=(25*scale, 12*scale))
mpl.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
    'font.family': 'Times New Roman',
    'font.size': 65,
    'axes.unicode_minus': False,
    'mathtext.fontset': 'custom',
    'mathtext.rm': 'Times New Roman',
    'mathtext.it': 'Times New Roman:italic',
    'mathtext.bf': 'Times New Roman:bold',
    'axes.linewidth': 5,
    'xtick.major.size': 20,
    'xtick.major.width': 10,
    'xtick.minor.size': 10,
    'xtick.minor.width': 3,
    'ytick.minor.size': 10,
    'ytick.minor.width': 3,
    'ytick.major.size': 20,
    'ytick.major.width': 5,
    "markers.fillstyle" :  'left',
    "lines.markersize" : 20,
    "lines.markeredgewidth" : 2,
})

bar_width = 0.175
shrink = 0.2
colors = ["#7CBEEC", "#7EB57E", "#FF6F45", "#3274A1"]
labels = ['success rate (%)', 'in_constraints_probs (%)', 'runtime']
indexs = [i for i in np.arange(0, len(data1024shots.index.values))]
for i in range(0, len(indexs) - 1, 3):
    indexs[i] += shrink
    indexs[i + 2] -= shrink

for i, metric in enumerate(["best_solution_probs", 'in_constraints_probs', 'runtime']):
    axes = plt.axes([0, 0 + i * 0.5, 0.8, 0.37])
    ylabel = labels[i]
    ymetric = metric

    axes.bar([index - 1.5 * bar_width for index in indexs], data1024shots[ymetric]['penalty'], width=bar_width, label="penalty", color=colors[0])
    axes.bar([index - 0.5 * bar_width for index in indexs], data1024shots[ymetric]['cyclic'], width=bar_width, label="cyclic", color=colors[1])
    axes.bar([index + 0.5 * bar_width for index in indexs], data1024shots[ymetric]['HEA'], width=bar_width, label="HEA", color=colors[2])
    axes.bar([index + 1.5 * bar_width for index in indexs], data1024shots[ymetric]['commute'], width=bar_width, label="QAcOA-CH", color=colors[3])
    axes2 = axes.twinx()
    improvement = data1024shots[ymetric]['commute'] / data1024shots[ymetric][['penalty', 'cyclic', 'HEA']].mean(axis=1)
    axes2.plot(indexs, improvement, color='k', marker='^', linestyle='--', markersize=15, markeredgewidth=3, label='Improvement')
    
    
    if metric == "best_solution_probs":
        axes.set_yticks(range(0, 8, 1), minor=True)
        axes.set_yticks(range(0, 8, 5), minor=False)
        axes2.set_yticks([i / 10 for i in range(0, 40, 2)], minor=True)
        axes2.set_yticks(range(0, 4, 1), minor=False)
        # axes2.grid(True, which='both', linestyle='--', linewidth=1)
        # exit()
        # improvement = data1024shots[ymetric][['penalty', 'cyclic', 'HEA']].apply(lambda x: data1024shots[ymetric]['commute'] / x)
        # axes2.boxplot(improvement.T, positions= layers, showfliers=False, widths=bar_width, patch_artist=True, 
        #               boxprops=dict(facecolor='white', linewidth=2), 
        #               medianprops=dict(color="red", linewidth=2), 
        #               whiskerprops=dict(color="black", linewidth=2), 
        #               capprops=dict(color="black", linewidth=2))
        axes2.set_ylabel(r"improvement ($\times$)")
    elif metric == "in_constraints_probs":
        axes.set_yticks(range(0, 51, 5), minor=True)
        axes.set_yticks(range(0, 51, 10), minor=False)
        axes2.set_yticks(range(0, 6, 1), minor=True)
        axes2.set_yticks(range(0, 6, 2), minor=False)
        axes2.set_ylabel(r"improvement ($\times$)")

    elif metric == "runtime":
        axes.set_yticks(range(0, 29, 5), minor=True)
        axes.set_yticks(range(0, 29, 10), minor=False)
        axes2.set_yticks([i / 4 for i in range(0, 6, 1)], minor=True)
        axes2.set_yticks([i / 4 for i in range(0, 6, 1)], minor=False)
#         improvement = data1024shots[ymetric][['penalty', 'cyclic', 'HEA']].apply(lambda x: x / data1024shots[ymetric]['commute'])
#         # axes2.boxplot(improvement.T, positions=layers+1*bar_width, showfliers=False, widths=bar_width, patch_artist=True,
#         #               boxprops=dict(facecolor='white', linewidth=2), 
#         #               medianprops=dict(color="red", linewidth=2), 
#         #               whiskerprops=dict(color="black", linewidth=2), capprops=dict(color="black", linewidth=2))
#         axes2.legend(frameon=False, loc='upper right', prop={'size': 60})
        axes2.set_ylabel(r"speedup ($\times$)")
    # axes.tick_params(axis="x", which="major", width=5, length=20)
    index = np.append(np.arange(1, len(data1024shots.index.values), 3), len(data1024shots.index.values) - 1)
    label = list(data1024shots.index.levels[0].values) + ['Avg']
    axes.set_xticks(index, minor=False)
    axes.set_xticklabels(label, minor=False)
    axes.set_xticks(indexs, minor=True)
    axes.grid(axis="y", color="gray", linestyle="--", linewidth=1)
    axes.set_facecolor("#fffef877")
    axes.set_ylabel(ylabel)
    # axes.set_yscale('log')
axes.legend(frameon=False, bbox_to_anchor=(0.5, 1.05), loc='lower center', ncol=4, prop={'size': 60})
# plt.tight_layout()
plt.savefig("qpu.svg", bbox_inches="tight")
```
