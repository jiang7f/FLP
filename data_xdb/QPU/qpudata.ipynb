```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
data = pd.read_csv('realdata.csv')
data.drop(columns=['pbid'], inplace=True)
data
```

```python
latencydf = pd.read_csv('_latency&transpile_time.csv')
latencydf
```

```python
qpu_data =data.merge(latencydf, on=['pkid','layers','method','backend'])
qpu_data.latency = qpu_data.latency
qpu_data
```

```python
qpu_data['runtime'] = qpu_data.latency*qpu_data.shots*qpu_data.iteration_count/10**9+ qpu_data.transpile_time
qpu_data
```

```python
probs = ['F1','G1','K1']
qpu_data.pkid = qpu_data.pkid.apply(lambda x: probs[x])
qpu_data
```

```python

data2draw = qpu_data.pivot(index =['backend','pkid','shots'], columns='method', values=["ARG",'in_constraints_probs','best_solution_probs','iteration_count','latency','transpile_time','runtime'])
data2draw
```

```python
data2draw.groupby(['backend','shots']).mean()
```

## 展示真机上求解结果与求解时间

```python
data1024shots = data2draw.xs(1024, level='shots')
```

```python
data1024shots
```

```python
data1024shots.index.levels[0]
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
scale = 1.05
fig= plt.figure(figsize=(25*scale, 12*scale))
mpl.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
    'font.family': 'Times New Roman',
    'font.size': 65,
    'axes.unicode_minus': False,
    'mathtext.fontset': 'custom',
    'mathtext.rm': 'Times New Roman',
    'mathtext.it': 'Times New Roman:italic',
    'mathtext.bf': 'Times New Roman:bold',
    'axes.linewidth': 5,
    'xtick.major.size': 20,
    'xtick.major.width': 10,
    'ytick.minor.size': 10,
    'ytick.minor.width': 3,
    'ytick.major.size': 20,
    'ytick.major.width': 5,
    "markers.fillstyle" :  'left' , ## default
    "lines.markersize" : 20,
    "lines.markeredgewidth" : 2,
})
bar_width =  0.15
colors = ["#7CBEEC", "#7EB57E", "#FF6F45", "#3274A1"]
labels = ['success rate (%)', 'in_constraints_probs (%)']

for i, metric in enumerate( ["best_solution_probs",'in_constraints_probs']):
    axes = plt.axes([0, 0+i*0.5, 0.8, 0.37])
    ylabel = labels[i]
    layers = np.arange(1,len(data1024shots.index.values)+1)

    ymetric = metric# best_solution_probs
    axes.bar(layers-1*bar_width,data1024shots[ymetric]['penalty'], width=bar_width,
            label="penalty", color=colors[0])
    axes.bar(layers+0*bar_width,data1024shots[ymetric]['cyclic'], width=bar_width,
            label="cyclic", color=colors[1])
    axes.bar(layers+1*bar_width,data1024shots[ymetric]['HEA'], width=bar_width,
            label="HEA", color=colors[2])
    axes.bar(layers+2*bar_width,data1024shots[ymetric]['commute'], width=bar_width,
            label="QAcOA-CH", color=colors[3])
    
    if i == 0:
        # axes.set_ylim(0,40)
        axes.set_yticks(range(0,15,1), minor=True)
        axes.set_yticks(range(0,15,5), minor=False)
        axes2 = axes.twinx()
        # axes2.set_ylim(0,50)
        # axes.set_ylim(0,50)
        axes2.set_yticks(range(0,50,5), minor=True)
        axes2.set_yticks(range(0,50,20), minor=False)
        improvement = data1024shots[ymetric][['penalty','cyclic','HEA']].apply(lambda x:data1024shots[ymetric]['commute']/x)
        axes2.boxplot(improvement.T, positions=layers+1*bar_width, showfliers=False, widths= bar_width, patch_artist=True, 
                      boxprops=dict(facecolor='white', linewidth=2), 
                      medianprops=dict(color="red", linewidth=2), 
                      whiskerprops=dict(color="black", linewidth=2), 
                      capprops=dict(color="black", linewidth=2))
        # axes2.set_ylim(0,40)
        axes2.set_ylabel(r"improvement ($\times$)")
        # axes2.set_xticks(None,minor=True)
        # axes.set_yticks(range(0,55,5), minor=True)
        # axes.set_yticks(range(0,55,10), minor=False)
        # axes.set_yscale("log")
    else:
        # axes.set_ylim(0,90)
        axes.set_yticks(range(0,41,5), minor=True)
        axes.set_yticks(range(0,41,10), minor=False)
        
        axes2 = axes.twinx()
        # axes2.set_ylim(0,9)
        # axes.set_ylim(0,50)
        axes2.set_yticks(range(0,9,1), minor=True)
        axes2.set_yticks(range(0,9,4), minor=False)
        improvement = data1024shots[ymetric][['penalty','cyclic','HEA']].apply(lambda x: x/data1024shots[ymetric]['commute'])
        axes2.boxplot(improvement.T, positions=layers+1*bar_width, showfliers=False, widths= bar_width, patch_artist=True,
                      boxprops=dict(facecolor='white', linewidth=2), 
                      medianprops=dict(color="red", linewidth=2), 
                      whiskerprops=dict(color="black", linewidth=2), capprops=dict(color="black", linewidth=2))
        # axes2.set_ylim(0,40)
        axes2.legend(frameon=False, loc='upper right', prop={'size': 60})
        axes2.set_ylabel(r"speedup ($\times$)")
#     axes.set_yticks(range(0,101,20), minor=False)
    axes.tick_params(axis="x", which="major", width=5, length=20)
    axes.set_xticks(np.arange(1,len(data1024shots.index.values)+1)+0.5*bar_width,minor=True)
    # axes.setdefault(key, value)
    axes.set_xticks(np.arange(1,len(data1024shots.index.values)+1,3)+1+0.5*bar_width,minor=False)
    axes.set_xticklabels([x.replace('Fake','') for x in data1024shots.index.levels[0].values], minor=False)
    # axes.set_xticklabels()
    # axes.set_xticklabels([prob_lab[idx] for idx in mean_commute.variables])
    axes.grid(axis="y", color="gray", linestyle="--", linewidth=1)
    ### set bakground color
    axes.set_facecolor("#fffef877")
    # axes.set_title("Comparison of QAcOA-CH with other algorithms")
    # axes.set_xlabel("#layer")
    axes.set_ylabel(ylabel)
#     axes.set_yscale("log")
axes.legend(frameon=False, bbox_to_anchor=(0.5, 1.05), loc='lower center', ncol=4, prop={'size': 60})
plt.tight_layout()
plt.savefig("qpuruntimeandaccuracy.svg", bbox_inches="tight")
```
